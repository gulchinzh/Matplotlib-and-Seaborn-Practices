# 4. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"
# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
labels=["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
```

```python
g = df.plot.bar(figsize=(9,6), rot=60)

for i in list(g.patches):
    print(i)
```

```python
g = df.plot.bar(figsize=(9,6), rot=60)

g.set_xticklabels(labels)

for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.03, p.get_height()+2))
    
    # annotate --> ilk parametre text --> buraya patch'lerin height değerlerini giremek istiyoruz. bunu get_height ile yapıyoruz.
    #              ikinci parametre --> xy --> girilecek text'in konumunu belirliyoruz. bunun içindeki ikinci değer y, 
    #              yani yükseklik yerine get_height deseydik tam yükseklik değeri üzerine text'i koyacaktı. biraz üzerinde olması için bu değerlere + 5 ekledik.
```
